```python
import matplotlib.pyplot as plt
```

```python
# Due to go-deep library limitations, it was mandatory to copy-paste results by hand
epochs = range(10)
centralized_protocole_performances = [0.59, 0.70, 0.83, 0.79, 0.83, 0.83, 0.80, 0.85, 0.83, 0.84]
```

```python
fig, ax = plt.subplots(constrained_layout=True)
fig.suptitle("Centralized protocole performance", fontsize=16)
ax.plot(epochs, centralized_protocole_performances, label='Test accuracy')

ax.set_title("1 layer | 32 neurons | ReLU activation")
plt.xlabel("epochs")
plt.ylabel("accuracy")
ax.legend()
plt.show()
```

```python
worker1 = [0.89, 0.92]
worker2 = [0.88, 0.92]
master = [0.73, 0.93]
epochs = range(2)
```

```python
fig, ax = plt.subplots(constrained_layout=True)
fig.suptitle("Federated protocole accuracy performance", fontsize=16)
ax.set_title("1 layer | 32 neurons | ReLU activation")

ax.plot(epochs, worker1, label='worker1')
ax.plot(epochs, worker2, label='worker2')
ax.plot(epochs, master, label='master')


plt.xlabel("epochs")
plt.ylabel("accuracy")
ax.legend()
plt.show()
```
